### Normalize

```python
from sklearn.preprocessing import StandardScaler
df2 = df1.drop(['country'], axis=1)
sclr = StandardScaler()
feature_norm = sclr.fit_transform(df2)
```

## PCA

```python
cov_matrix= np.cov(feature_norm.T)
eign_vals , eign_vecs = np.linalg.eig(cov_matrix)
print ("Eignvaies : \n",eign_vals)
```

```python
sum_val= sum(eign_vals)
var_exp =[(i/sum_val) for i in sorted (eign_vals , reverse=True)]
cum_var_exp =np.cumsum (var_exp)


plt.figure(figsize=(10,8))
plt.bar(range(1,10) , var_exp , alpha= 0.5 , align='center' , label='var_exp')
plt.step(range(1,10),cum_var_exp , where='mid' , label='cumulative var_exp')
plt.ylabel('Explained variance ratio')
plt.xlabel('priciple component')
plt.legend(loc= 'best')
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(feature_norm)
pca_=pca.transform(feature_norm)
DF = pd.DataFrame(pca.transform(feature_norm), columns=(["PCA1","PCA2"]))
DF
```

```python
from fcmeans import FCM
from sklearn.metrics import davies_bouldin_score
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import silhouette_score

fpcs = []
num_cluster=[2,3,4,5]
label=[]
cntr=[]
coe =[]
silhoutte=[]
harabasz=[]
davies=[]

for i in range(len(num_cluster)):
    fcm = FCM(n_clusters = num_cluster[i])
    fcm.fit(pca_)
    fcm.predict(pca_)
    fcm_coe=fcm.partition_coefficient
    fcm_centers = fcm.centers
    fcm_labels=fcm.predict(pca_)
    label.append(fcm_labels)
    #cntr.append(fcm_centers)
    coe.append(fcm_coe)
    score1 = silhouette_score(pca_, fcm_labels)
    score2 = calinski_harabasz_score(pca_, fcm_labels)
    score3 = davies_bouldin_score(pca_, fcm_labels)
    
    silhoutte.append(score1)   
    harabasz.append(score2)
    davies.append(score3)

dict1={'num_cluster':num_cluster, 'silhoutte_score':silhoutte }
dict2={'num_cluster':num_cluster, 'harabasz_score':harabasz }
dict3={'num_cluster':num_cluster, 'davies_score':davies }

evsilhoutte_scr= pd.DataFrame(dict1)
evharabasz_scr= pd.DataFrame(dict2)    
evdavies_scr= pd.DataFrame(dict3)
```

```python
# Create 2x2 sub plots

fig, axs = plt.subplots(2, 2,figsize=(15,12))

axs[0, 0].plot(num_cluster,silhoutte,'tab:red')
axs[0, 0].set_title('silhoutte coeffcient',color='black')
axs[0, 0].tick_params(axis='x', colors='black')
axs[0, 0].tick_params(axis='y', colors='black')


axs[0, 1].plot(num_cluster,harabasz,'tab:orange')
axs[0, 1].set_title('calinski_harabasz coeffcient',color='black')
axs[0, 1].tick_params(axis='x', colors='black')
axs[0, 1].tick_params(axis='y', colors='black')



axs[1, 0].plot(num_cluster,davies,'tab:green')
axs[1, 0].set_title('davies_bouldin coeffcient',color='black')
axs[1, 0].tick_params(axis='x', colors='black')
axs[1, 0].tick_params(axis='y', colors='black')



axs[1, 1].plot(num_cluster,coe)
axs[1, 1].set_title('FCM partition coefficient',color='black')
axs[1, 1].tick_params(axis='x', colors='black')
axs[1, 1].tick_params(axis='y', colors='black')


for ax in axs.flat:
    ax.set(xlabel='Number of cluster', ylabel='score', xticks=[2,3,4,5])
    
    
    

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering 
from sklearn.metrics import davies_bouldin_score
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import silhouette_score


num_cluster=[2,3,4,5]
silhoutte=[]
harabasz=[]
davies=[]

for i in range(len(num_cluster)):
    h_cluster = AgglomerativeClustering(n_clusters = num_cluster[i], affinity = 'euclidean', linkage ='average')
    h_cluster = h_cluster.fit(pca_)
    h_cluster_pred= h_cluster.fit_predict(pca_)
    
    score1 = silhouette_score(pca_, h_cluster_pred)
    score2 = calinski_harabasz_score(pca_, h_cluster_pred)
    score3 = davies_bouldin_score(pca_, h_cluster_pred)
    
    silhoutte.append(score1)   
    harabasz.append(score2)
    davies.append(score3)

dict1={'num_cluster':num_cluster, 'silhoutte_score':silhoutte }
dict2={'num_cluster':num_cluster, 'harabasz_score':harabasz }
dict3={'num_cluster':num_cluster, 'davies_score':davies }

evsilhoutte_scr= pd.DataFrame(dict1)
evharabasz_scr= pd.DataFrame(dict2)    
evdavies_scr= pd.DataFrame(dict3)
```
